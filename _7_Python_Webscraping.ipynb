```python
import newspaper
```

# Build source objects to the website





```python
paper = newspaper.build('https://towardsdatascience.com/')
```

# Extracting aricles



```python
print(paper.size())
```

```python
for article in paper.articles:
  print(article.url)
```

# Extracting Source category 






```python
for category in paper.category_urls():
  print(category)
```

# Extracting Source brand & description

```python
print("brand:", paper.brand) 
print("Description:",paper.description)
```

# Let's extract the first article

```python
first_article = paper.articles[0]
first_article.download()
```

```python
print(first_article.html)
```

# Analyzing the article

```python
first_article.parse()
```

```python
print(first_article.text)
```

```python
print(first_article.title)
```

```python
print(first_article.authors)
```

```python
print(first_article.top_image)
```

```python
print(first_article.images)
```

```python
print(first_article.movies)
```

# Natural Language Processing on the article

```python
import nltk
nltk.download('punkt')
first_article.nlp() #You must have called both download() and parse() on the article before calling nlp()
```

```python
print(first_article.summary)
```

```python
print(first_article.keywords)
```
